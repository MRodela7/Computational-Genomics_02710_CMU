```python
import math

def viterbi(obs, states, start_p, trans_p, emit_p):
    V = [{}]
    path = {}

    # Initialize base cases (t=0)
    for y in states:
        V[0][y] = math.log(start_p[y]) + emit_p[y][obs[0][0]]
        path[y] = [y]

    # Run Viterbi for t > 0
    for t in range(1, len(obs)):
        V.append({})
        newpath = {}

        for y in states:
            (prob, state) = max((V[t-1][y0] + trans_p[y0, y] + emit_p[y][obs[t][i+1]], y0) for i,y0 in enumerate(states))
            V[t][y] = prob
            newpath[y] = path[state] + [y]

        # Don't need to remember the old paths
        path = newpath

    (prob, state) = max((V[-1][y], y) for y in states)
    return prob, path[state]

import numpy as np
# Test the function with the given arguments
obs = [['G', 'C', 'A', 'T', 'G', 'T', 'A', 'T', 'C', 'A'], ['G', 'C', 'A', 'T', 'G', 'T', 'A', 'T', 'C', 'A'],['C', 'T', 'A', 'G', 'A', 'G', 'T', 'A', 'G', 'C'], ['C', 'T', 'A', 'G', 'A', 'G', 'T', 'A', 'G', 'C']]
print("len of obs:", len(obs), np.asarray(obs).shape)
print("len of output list:", len(output_list), np.asarray(output_list).shape)
states = ['V', 'W']
start_p = {'V': 0.5, 'W': 0.5}
trans_p = {('V', 'W'): math.log(0.25), ('W', 'V'): math.log(0.25), ('W', 'W'): math.log(0.75), ('V', 'V'): math.log(0.75), ('V', 'A'): math.log(0.30), ('V', 'C'): math.log(0.20), ('V', 'G'): math.log(0.20), ('V', 'T'): math.log(0.30), ('W', 'A'): math.log(0.20), ('W', 'C'): math.log(0.30), ('W', 'G'): math.log(0.30), ('W', 'T'): math.log(0.20)}
emit_p = {'V': {'A': math.log(0.10), 'C': math.log(0.40), 'G': math.log(0.40), 'T': math.log(0.10)}, 'W': {'A': math.log(0.40), 'C': math.log(0.10), 'G': math.log(0.10), 'T': math.log(0.40)}}


# prob, path = viterbi(output_list, states, start_p, trans_p, emit_p)
prob, path = viterbi(obs, states, start_p, trans_p, emit_p)
print("Most likely path: ", path)
print("Probability of the path: ", prob)
```

```python
import math
import numpy as np
from scipy.special import logsumexp

def forward(obs, states, start_p, trans_p, emit_p):
    alpha = [{}]
    for state in states:
        alpha[0][state] = np.log(start_p[state]) + emit_p[state].get(obs[0], -np.inf)
    for t in range(1, len(obs)):
        alpha.append({})
        for state in states:
            if obs[t] not in emit_p[state]:
                emit_p[state][obs[t]] = -np.inf
            alpha[t][state] = emit_p[state][obs[t]] + logsumexp([alpha[t-1][prev_state] + trans_p[(prev_state, state)] for prev_state in states])
    return alpha

def backward(obs, states, start_p, trans_p, emit_p):
    beta = [{} for _ in range(len(obs))]
    # Initialize base case (t = T)
    for state in states:
        beta[len(obs)-1][state] = 1
    # Run backward for t < T
    for t in range(len(obs)-2, -1, -1):
        for state in states:
            beta[t][state] = sum(trans_p[(state, next_state)] + emit_p[next_state][obs[t+1]] + beta[t+1][next_state] for next_state in states)
    return beta

def viterbi(obs, states, start_p, trans_p, emit_p):
    V = [{}]
    for state in states:
        V[0][state] = start_p[state] + emit_p[state][obs[0]]
    path = {state:[state] for state in states}
    for t in range(1, len(obs)):
        V.append({})
        newpath = {}
        for state in states:
            (prob, prev_state) = max((V[t-1][prev_state] + trans_p[(prev_state, state)] + emit_p[state][obs[t]], prev_state) for prev_state in states)
            V[t][state] = prob
            newpath[state] = path[prev_state] + [state]
        path = newpath
    (prob, state) = max((V[t][state], state) for state in states)
    return (prob, path[state])


def train_viterbi(obs, states, start_p, trans_p, emit_p, max_iterations=100, convergence_threshold=1e-5):
    for iteration in range(max_iterations):
        old_trans_p = trans_p.copy()  # make a copy of trans_p to check for convergence later
        for seq in obs:
            alpha = forward(seq, states, start_p, trans_p, emit_p)
            beta = backward(seq, states, start_p, trans_p, emit_p)
            gamma = np.exp(alpha + beta - logsumexp(alpha[-1].values()))  # compute the gamma matrix
            
            # re-estimate start_p
            start_p = gamma[0]
            
            # re-estimate emit_p
            for state in states:
                emit_p[state] = np.exp(logsumexp(gamma[t][state] for t in range(len(seq))) - logsumexp(gamma[t][state] for t in range(len(seq))))
            
            # re-estimate trans_p
            for i in range(len(states)):
                for j in range(len(states)):
                    trans_p[(states[i], states[j])] = np.exp(logsumexp(alpha[t][states[i]] + beta[t+1][states[j]] + emit_p[states[j]][seq[t+1]] for t in range(len(seq)-1)) - logsumexp(gamma[t][states[i]] for t in range(len(seq)-1)))
        
        # check for convergence of trans_p
        diff = np.sum(np.abs(trans_p - old_trans_p))
        if diff < convergence_threshold:
            break
            
    return start_p, trans_p, emit_p



#Samll dataset
# obs = [['G', 'C', 'A', 'T', 'G', 'T', 'A', 'T', 'C', 'A'], ['C', 'T', 'A', 'G', 'A', 'G', 'T', 'A', 'G', 'C']]

states = ['V', 'W']
start_p = {'V': 0.5, 'W': 0.5}
trans_p = {('V', 'W'): math.log(0.25), ('W', 'V'): math.log(0.25), ('W', 'W'): math.log(0.75), ('V', 'V'): math.log(0.75), ('V', 'A'): math.log(0.30), ('V', 'C'): math.log(0.20), ('V', 'G'): math.log(0.20), ('V', 'T'): math.log(0.30), ('W', 'A'): math.log(0.20), ('W', 'C'): math.log(0.30), ('W', 'G'): math.log(0.30), ('W', 'T'): math.log(0.20)}
emit_p = {'V': {'A': math.log(0.10), 'C': math.log(0.40), 'G': math.log(0.40), 'T': math.log(0.10)}, 'W': {'A': math.log(0.40), 'C': math.log(0.10), 'G': math.log(0.10), 'T': math.log(0.40)}}

#load training obs data
obs = []
with open('werepyre-train.fasta', "r") as f:
  for line in f:
    line = line.split()
    if '>' in line[0]:
      continue
    else:obs.append(line)
output_list = [[char for char in string[0]] for string in obs]
print(output_list)

#Train Viterbi and get the path
start_p, trans_p, emit_p =train_viterbi(output_list, states, start_p, trans_p, emit_p)
probPath = viterbi(output_list, states, start_p, trans_p, emit_p)

#Load test Data
test_obs = []
with open("werepyre-test.fasta", "r") as f:
  for line in f:
    line = line.split()
    if '>' in line[0]:
      continue
    else:test_obs.append(line)
output_list_test = [[char for char in string[0]] for string in test_obs]

#Open test annotation file:
test_obs_annot = []
with open("werepyre-test-annotation.fasta", "r") as f:
  for line in f:
    line = line.split()
    if '>' in line[0]:
      continue
    else:test_obs.append(line)
output_list_anno = [[char for char in string[0]] for string in test_obs_annot]

#check with test value probabilities
states = ['V', 'W']

trans_p = {}
emit_p = {}

alphabet = ['A', 'C', 'G', 'T']

trans_p = {}
emit_p = {}

trans_p[('V', 'W')] = math.log(0.010)
trans_p[('W', 'V')] = math.log(0.014)
trans_p[('W', 'W')] = math.log(0.986)
trans_p[('V', 'V')] = math.log(0.99)


emit_p[('V', 'A')] = math.log(0.35)
emit_p[('V', 'C')] = math.log(0.15)
emit_p[('V', 'G')] = math.log(0.15)
emit_p[('V', 'T')] = math.log(0.35)

emit_p[('W', 'A')] = math.log(0.175)
emit_p[('W', 'C')] = math.log(0.325)
emit_p[('W', 'G')] = math.log(0.325)
emit_p[('W', 'T')] = math.log(0.175)

test_true = viterbi(output_list_test, states, trans_p, emit_p)

#Calculate accuracy
matched = 0
mismatched = 0
for i in range(len(test_true)):
    if test_true[i] == test_annotation[i]:
        matched += 1
    else:
        mismatched += 1
accuracy = print(matched/len(test_true))
print("Accuracy:", accuracy)
```
